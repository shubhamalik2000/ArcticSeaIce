```python
def calculate_change_rate(start_value, end_value):
    return ((end_value - start_value) / start_value) * 100
```

```python
decade_groups = dsep.groupby(df.index // 10 * 10)
```

```python
decade_groups = dsep.groupby(dsep.index + (10 - dsep.index % 10) % 10)
```

```python
decade_groups
```

```python
decade_change_rates = pd.DataFrame(columns=['Decade', 'Change Rate'])
```

```python
for decade, group in decade_groups:
    start_value = group.iloc[0]['September']  # September value at the beginning of the decade
    end_value = group.iloc[-1]['September']   # September value at the end of the decade
    change_rate = calculate_change_rate(start_value, end_value)
    new_data = {'Decade': [decade],
                'Change Rate': [change_rate]}
    # Convert the new data into a DataFrame
    new_row = pd.DataFrame(new_data)
    decade_change_rates = pd.concat([decade_change_rates, new_row], ignore_index=True)
```

```python
# Given data
present_area = 4.371  # in million sq kms
present_rate = -0.1228  # Decadal change rate (-12.28% expressed as decimal)

# Convert the percentage rate to a decimal
rate_decimal = present_rate / 100

# Function to calculate time to extinction
def time_to_extinction(initial_area, rate):
    # Calculate time to extinction using exponential decay formula
    time = np.log(1 / initial_area) / rate
    return time

# Calculate time to extinction
time_to_extinct = time_to_extinction(present_area, rate_decimal)

print("Time to extinction (in years):", time_to_extinct)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM
```

```python
scaler = MinMaxScaler(feature_range = (0,1))
dataset = scaler.fit_transform(dataset)
```

```python
dataset = pd.DataFrame(dataset, index = dsep.index, columns = ['Scaled_value'])
```

```python
def create_dataset(dataset, look_back = 1):
    dataX, dataY = [], []
    for i in range(len(dataset) - 1):
        a = dataset[i:(i+look_back),0]
        dataX.append(a)
        dataY.append(dataset[i + look_back, 0])
    return np.array(dataX), np.array(dataY)
```

```python
'''
so the thing is we are creating two sets here, X and y. So as to transform this
into something more like f(X) = y. 
Beacuse, when predicting for day n, we need to have data for upto day n-1
'''
```

```python
trainA = np.array(train)
```

```python
trainX, trainY = create_dataset(trainA)
```

```python
trainX
```

```python
trainX.shape
```

```python
trainY.shape
```

```python
test.head()
```

```python
indext1 = test.index.min()
print(indext1)
indext2 = test.index.max()
print(indext2)
```

```python
testA = np.array(test)
```

```python
testX, testY = create_dataset(testA)
```

```python
testX.shape
```

```python
testY.shape
```

```python
trainX = np.reshape(trainX, (trainX.shape[0], 1, trainX.shape[1]))
testX = np.reshape(testX, (testX.shape[0], 1, testX.shape[1]))
```

```python
trainX.shape
```

```python
testX.shape
```

```python
# create and fit the LSTM model 

model = Sequential()
model.add(LSTM(4, input_shape = (1,1)))
model.add(Dense(1))
model.compile(loss = 'mean_squared_error', optimizer = 'adam')
model.fit(trainX, trainY, epochs = 100, batch_size = 1, verbose = 2)
```

```python
trainPredict = model.predict(trainX)
testPredict = model.predict(testX)
```

```python
# invert predictions

trainPredict = scaler.inverse_transform(trainPredict)
trainY = scaler.inverse_transform([trainY])
testPredict = scaler.inverse_transform(testPredict)
testY = scaler.inverse_transform([testY])
```

```python
# calculate the root mean squared error

trainScore = np.sqrt(mean_squared_error(trainY[0], trainPredict[:,0]))
print('Train score: %.2f RMSE' %(trainScore))
```

```python
rmse_lstm = np.sqrt(mean_squared_error(testY[0], testPredict[:,0]))
print('Test score: %.2f RMSE' %(rmse_lstm))
```
